### ONE HOT Encoding -- Encoding Nominal Data

```python
cars=pd.read_csv("test-data.csv")
```

```python
cars.sample(2)
```

```python
cars.set_index("Unnamed: 0")
```

```python
cars["Location"].unique()
```

```python
cars.head(2)
```

```python
cars=cars.iloc[:,1:7]
```

```python
cars.sample(2)
```

```python
cars["Fuel_Type"].unique()
```

```python
cars["Name"].value_counts()
```

```python
cars['brand'] = cars['Name'].str.split().str[0]
```

```python
cars.drop("Name", axis=1, inplace=True)
```

```python
cars.head()
```

```python
cars["brand"].unique()
```

```python
cars["brand"]=cars["brand"].replace("ISUZU","Isuzu")
```

```python
cars["brand"].value_counts()
```

```python
cars["Transmission"].value_counts()
```

```python
cars["Location"].value_counts()
```

### 1. ONE-HOT Encoding using Pandas

```python
pd.get_dummies(cars, columns=["Fuel_Type", "Transmission"], dtype=int)
```

### 2. K-1 Encoding

```python
pd.get_dummies(cars, columns=["Fuel_Type", "Transmission"], dtype=int, drop_first=True)
```

### 4. ONE-HOT Encoding with Top categories

```python
counts = cars['brand'].value_counts()
```

```python
cars['brand'].nunique()
```

```python
threshold=42
```

```python
repl=counts[counts >= threshold].index
```

```python
pd.get_dummies(cars['brand'].replace(repl, "uncommon"),dtype='int', drop_first=True).sample(10)
```
