## Bivariate Analysis

```python
import seaborn as sns
import pandas as pd
```

### A. Scatter Plot (Numerical-Numerical)

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.isnull().sum()
```

```python
sns.scatterplot(tips, x="total_bill", y="tip", hue='sex', style=tips['smoker'], size=tips['size'])
```

```python
sns.barplot(tips, x="smoker", y="tip")
```

### B. Bar Plot (Numerical-Categorical)

```python
titan = pd.read_csv('tested.csv')
```

```python
titan.head()
```

```python
sns.barplot(titan, x='Pclass', y='Fare', hue='Sex')
```

### C. Box Plot (Numerical-Categorical)

```python
sns.boxplot(titan, x='Sex', y='Age', hue='Embarked')
```

```python
print('In the titan dataset, all the males died and all the females survived.')
print('Miraculous dataset no!!!')
```

### D. Distplot

```python
sns.distplot(titan[titan['Survived']==0]['Age'], hist=False)
sns.distplot(titan[titan['Survived']==1]['Age'], hist=False)
```

```python
titan[titan['Age'] < 0]['Age']
```

### E. HeatMap (Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titan['Pclass'], titan['Survived']))
```

```python
import matplotlib.pyplot as plt
titan.groupby('Embarked')['Survived'].mean().plot(kind='bar')

plt.ylabel('Survived')
plt.title('Survival Rate per class')
```

### F. ClusterMap (Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titan['Parch'], titan['Survived']))
```

### G. Pairplot

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

### H. LinePlot (Numerical-Numerical)

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
 sns.lineplot(new, x='year', y='passengers')
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year', observed=False))
```

```python
sns.clustermap(flights.pivot_table(values='passengers', index='month', columns='year', observed=False))
```
